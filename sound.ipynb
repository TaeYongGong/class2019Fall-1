```python
from matplotlib import pyplot as plt #matplotlib에서 pyplot // from과 똑같은 역할 : import matplotlib.pyplot as plt  
from mpl_toolkits.mplot3d import Axes3D
from mpl_toolkits.axes_grid1 import make_axes_locatable
import IPython.display as ipd #ipython 내 display라는 함수를 ipd로 가져옴
import numpy as np
%matplotlib notebook
from scipy.signal import lfilter
```

```python
# parameter setting ->  밑에만 고쳐주면 결과값이 달라지기 때문에 parameter라고 부른다. 
amp = 1 # range [0.0, 1.0]
sr = 10000 # 얼마나 고 해상도 / 1초동안 어느정도의 정보를 담을 것이고 그것으로 표현할 것인가? 
dur = 0.5 # in seconds
freq = 100.0 # sine frequency, Hz // 1초에 반복이 몇번 되는가? 
```

```python
#t =  0.0001 0.0002 0.0003 .... 0.5000
# t을 만들어주는 이유 : 오일러 공식 -> theta에 들어가는 것은 무조건 각도값만 들어가야 한다. 
# but, 시간이 없으면 소리의 실체값을 만들어 줄 수가 없음 따라서 시간이라는 개념을 먼저 만들어줘야 함 
# sr와 관련 / 이것을 쉽게 만드는 방법은 무엇인가를 고민해야 한다. 
# 0.0001을 해줘야 한다 -> sr이 10000이기 때문
# 0.5000까지인 이유는 dur = 0.5이기 때문 
# time과 관련된 것은 dur 와 sr이다. 이 두가지만 있으면 시간이라는 개념을 만들어 낼 수 있다. 
```

```python
t = np.arange(1, sr+1) # +1을 하는 이유는 sr만 하면 9999까지만 되기 떄문 / sr만큼의 time tik을 만들어 줌
# 이건 그냥 숫자의 1차원 행렬의 나열이다.  
t
```

```python
t = np.arange(1, sr+1)/sr #1/10000초 부터 1초까지 만들어져서 시간을 표현가능하게 함. 왜냐하면 구간을 만들어 주기 때문이다. 즉, 분수로 
# 구분한 것이 시간이라는 개념으로 이어진다. 
# 구간이 곧 시간이된다.
t
# 1*1/10000 // 2*1/10000 .... 이어 진다. 
```

```python
# amplitude는 소리의 진폭을 더 크게 해줄 수 있음 
```

```python
# generate time 
t = np.arange(1, sr*dur+1)/sr
t
# 시간의 개념을 주기 위해서는 dur까지 포함 (dr=0.5이기 때문) / 1/10000<sampling rate만큼의 timetik을 만들어줌>
# 로 나뉜 구간을 5000까지만 해주고 이를 나열함 
# 이제 time을 기반으로 한 theta를 만들어 줄 수가 있음 
```

```python
# time이 있다고 다 되는 것은 아니다. 
# 각도 값을 time과 연동 시켜서 phasor로 바꿔줘야 한다. 
theta = t*2*np.pi*freq
# 즉 theta값은 
# t -> 시간의 행렬 / 벡터
# 2*np.pi = 2파이rad<360도> -> 하나의 주기다.<어떠한 그래프를 만드는 것이 아니라 x축의 값이다.> / np.pi = 1파이rad이다.<90도> 
# freq = 몇 바퀴를 만드느냐? -> 주기를 설정 -> frequency는 몇 바퀴를 도드냐를 정의해 준 것 
# 2*np.pi*freq 은 360도로 회전하는 것을 100번 반복하고 <사인그래프를 100번 > 
# 여기다가 *t를 해서 시간의 개념을 추가<구간을 나워주는 것>
# 그것을 0.5초동안 (200파아 rad을 10000개의 구간으로 나눈 뒤 5000번째 구간까지를 표현)
# time의 벡터의 사이즈와 theta의 벡터의 사이즈는 같다. // t는 arange이고 나머지는 곱하는 상수이기때문이다. 
# 즉, theta가 phase가 된다. -> 이는 phasor로 만들어 줄 수 있음 
theta
```

```python
# s = np.sin(theta) -> -2부터 2까지 만들어 주고 싶으면 2를 곱하면 된다. -> y축의 크기를 늘려줌 -> 진폭이라는게 설정하는 것은 결국 
s = amp*np.sin(theta) # -> 진폭이 2로 늘어남 // -1*x / 1*x로 x축을 기준으로 대칭적으로 커진다. 
```

```python
# generate signal by complex-phasor
c = np.exp(theta*1j)
c
# 실행 결과 하나 하나 전부 a+bi의 형태로 되어있다. 
# np.exp = 자연로그라고 생각하면 된다 <e> => e^theta*!j의 값 즉 NumPy의 np.exp(x) 함수는 밑(base)이 자연상수 e인 y=e^x지수함수로 변환
# 'cos(theta) + sin(theta)i 값'의 array로 구성되어 진다.
# 여기서 j는 허수 'i'다 / theta만 입력을 들어간다.<나머지는 고정이다> 
# -01 는 1/10 . -02는 1/10^2 -> 결국 이 소수점을 어디로 옮기느냐와 관련되어있다. 
# 이렇게 함으로서 정보의 양이 똑같음 (아무리 큰수든 작은수든 같은 정보량으로 표현된다. <개수가 same>)
# 행렬이 복소수값으로 채워져 있음<복소수의 벡터로 채워져 있음>
# 모든 수가 a+bi의 형태로 되어있구나 
```

```python
# 위의 c에도 전체에 amp.를 곱해주면 x축, y축, z축에 각각 amp만큼 비례적으로 늘어난다. 
amp = 2
c = amp*np.exp(theta*1j) 
c
```

```python
# 위에 것을 어떻게 plot할까가 중요한 문제 
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') # projection을 3d로 만들어준다<한 theta에 따라 표현되는 값이 3개 이기 때문이다.입력이 3개다> 
# 아무것도 없으면 2d다. 
ax.plot(t[0:1000], c.real[0:1000], c.imag[0:1000], '.') 
# 3차원 벡터가 된다. <3차원의 행렬> 이 모든 값은 0,1000개로 들어있는 값이 corresponding 한다 .
# 이미 c는 3개의 값으로 이뤄진 함수 <theta값에 t라는 개념이 들어가 있고 나머지는 그것의 결과값이다.>
# c.real이라는 것은 a의 값만 가져온다는 것<실수, cos부분> / '.'을 통해 c에서 real 부분을 가져온다.
# c.imag은 b의 값만 가져온다는 것<허수부분, sin부분> / '.'을 통해 c에서 imag부분을 가져온다.
ax.set_xlabel('time (s)') #시간
ax.set_ylabel('real') #실수
ax.set_zlabel('imag') #허수 -> theta의 값으로 달라지는
# 총 1000개의 점이 찍힌다. / theta값에 종속되어 있기 때문에 전부 -1에서 1사이의 값이다
# a, b의 값을 따도 따로 받아 와서 plot을 한다. 여기다가 t의 값도 가져와서 plot한다. 
```

```python
# generate samples, note conversion to float32 array
F0 = 100; Fend = int(sr/2); s = np.zeros(len(t)); 
amp = 10
#f0는 가장 낮은 frequency이다 / 우리는 배수로 올릴수 있는데 이것의 한계는 어디까지인가? => sr/2까지 가능하다. 
# fend -> 표현하는 제일 마지막 주파수 따라서 이는 sr/2가 되고 / 이 숫자는 소수점이기 때문에 int(반올림 정수 처리를 했다.)
# 여기서 s값은 가장 초기의 s값이다. 가장 처음 값은 0을 원소로 갖는, t만큼의 정보량을 가진 array이다. -> 그냥 s값을 0으로 정해주는 함수
for freq in range(F0, Fend+1, F0):
    theta = t * 2*np.pi * freq
    # 위에서 theta를 만들어 준 과정과 똑같음 -> f0 * 2파이 * t(벡터)이기 때문에 theta의 값도 벡터가 된다. 
    # freq = 몇 바퀴를 만드느냐? -> 주기를 설정 -> frequency는 몇 바퀴를 도드냐를 정의해 준 것
    tmp = amp * np.sin(theta)
    # 이를 sin에다 입력을 하고 amp도 적용해 준다 -> tmp도 하나의 벡터이다.
    s = s + tmp
    # 그것을 freq의 값을 높이면서 반복하는 과정 = s를 계속 더해서 업데이트해주는 과정 따라서 s의 최초값을 정의해 줘야함
    # 정보량이 같은 0으로 된 벡터 값을 s의 최초값으로 설정해 줘야 한다. 
    # for 루프 돌때 range 부분이 중요한데 이는 f0 = 100부터 fend까지<fend + 1> 이를 f0의 배수만큼 올려주기 위해서
    # <f0 = 100, f1 = 200, f2 = 300.... f49 = 5000>이 생긴다. (frequency의 값으로 theta에 집어넣진다.)
    # f0일때 부터, f49일때 까지의 tmp 행렬이 전부 합쳐져서 최종적으로 s라는 행렬을 이룬다.
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(t[0:1000], s[0:1000]); 
# t는 시간을 5000까지 나눈 구간에서 1000번째까지<x축>, s는 최종의 행렬에서 1000번째 값까지<y축> 그리고 s에는 이미 t라는 값이 포함되어 있음 
ax.set_xlabel('time (msec)')
ipd.Audio(s, rate=sr)
# 이를 왜 pulse train이라고 명명한 이유는 sin wave을 어떠한 틀에 넣어서 나온 정교한 소리의 그래프이기 때문
# 1항은 t = 0/5000, 1/5000, 2/5000.... 5000/50000 까지 전부 f0와 2파이를 곱해준다. 이를 f49까지 50번 시행한 것을 더한 행렬이 s임다
# 한마디로 5000개의 항으로 이루어진 벡터에서 t값에 따른 s값이 지정되어 있음 / t = 0 -> s의 첫번째 항, t = 1 -> s의 두번째 항 
```

```python
# 여태까지 우리는 누군가 만든 function을 이용했는데 이는 function을 만드는 방법이다. 
def hz2w(F, sr): #def로 함수 정의를 시작 / hz2w모양으로 hz2w의 변수를 F와 sr로 세팅해 준다. / F는 주파수, sr은 sampling rate
    NyFreq = sr/2; #NyFreq는 sampling rate의 1/2이다. 
    w = F/NyFreq *np.pi; # w는 (주파수)/(sampling rate의 1/2) 을 π랑 곱합 것이다. 
    return w # 이 부분이 출력이다. 

def resonance (srate, F, BW): # BW는 band with 이다.
    a2 = np.exp(-hz2w(BW,srate)) #hz2w(BW/sampling rate의 1/2 을 파이랑 곱합 것)를 불러온다. a2 = e^(-Bw/0.5sr*π)
    omega = F*2*np.pi/srate #omega = F*2*π/sr
    a1 = -2*np.sqrt(a2)*np.cos(omega) 
    #a1 = -2*[np.sqrt( e^(-Bw/0.5sr*π)) -> 각 항의 제곱근으로 변환 <4->2이런식으로>]*cos함수에(F*2*π/sr)을 대입한 결과 
    a = np.array([1, a1, a2]) 
    # -> a = (1, -2*√(e^(-Bw/0.5sr*π))*cos(F*2*π/sr), e^(-Bw/0.5sr*π))라는 벡터이다. 
    b = np.array([sum(a)])
    # -> 이러한 a의 값을 전부 더한 벡터이다. <1 + -2*√(e^(-Bw/0.5sr*π))*cos(F*2*π/srate) + e^(-Bw/0.5sr*π)>
    return a, b
```

```python
RG = 0  
# RG is the frequency of the Glottal Resonator -> 0이라는 주파수에다가 
BWG = 100  
# BWG is the bandwidth of the Glottal Resonator -> 대역폭의 설정<peak의 길이> -> 하나의 peak를 만들어줌
a, b=resonance(sr,RG,BWG) #a, b=resonance(10000, 0, 100)
#resonance sr = sampling rate / rg -> 산맥의 위치(peak의 위치)
# bwg -> 산맥이 얼마나 뾰족하냐 width(대역폭)가 크면 완만 / width(대역폭)가 작으면 뾰족 / 100정도변 완만 peak 
s = lfilter(b, a, s, axis=0) # s는 연속적으로 하기 위해서 계속 받아줘야한다. / 하나의 바퀴를 만들어주는 함수 
# a = (1, -2*√(e^(-1/50*π))*cos(0*2*π/10000), e^(-1/50*π))라는 벡터 / b는 이러한 a를 전부 더한 하나의 값으로 되어있는 벡터
ipd.Audio(s, rate=sr)
```

```python
# sr, s = wavfile.read('a.wav')
nSamp = len(s) # -> 5000
#s는 우리가 만든 wave<target wave>
# -> s의 길이를 알아야 그 phaosor만큼의 길이로 만들어 줄수 있다. (길이가 같아야지 inner product를 해줄 수 있다.)  
# 결국 비교하려믄 sample의 길이는 target wave와 같다. 
dur = nSamp / sr # 5000/10000 = 0.5 
t = np.linspace(1/sr, dur, nSamp)
# y = linspace(x1,x2)는 x1 -> 과 x2 사이에서 균일한 간격의 점 100개로 구성된 행 벡터를 반환합니다
# 여기서는 1/10000 부터, 0.5 = 5000/10000까지 점 5000개로 구성된 행 벡터 t를 뜻함 
# 1/10000, 2/10000, 3/10000 ....., 5000/10000 -> 벡터 
```

```python
nFFT = nSamp #nFFT는 sample 개수와 같다. -> 5000
amp = [];
for n in range(0,nFFT): # 샘플 개수만큼 for 루프를 돈다. -> 0부터 4999까지 -> 5000번 
    omega = 2*np.pi*n/nFFT # angular velocity / omega = w = 2π*(0,1,2, ....,4999 -> n의 값)/5000 까지 
    z = np.exp(omega*1j)**(np.arange(0,nSamp)) # -> (0,4999)까지  
    # n = 0 -> w=0, -> 1^(0,...4999) -> z값은 1이 5000개 있는 벡터값 -> 고려x 
    # n = 1 -> w=2π*1/5000 -> [e^(2/5000π*(0,1,....,4999))i]= -> (1,e^(2/5000π)i,e^(4/5000π)i, .... e^(2*4999/5000π)i)
    # e^(2n/5000π*(0,1,....,4999))i의 벡터
    # **는 지수라고 생각하면 된다. -> ex) **3은 세제곱이 된다. 
    # z -> complex phasor를 만드는 줄 -> z -> z1, .... z(n-1)까지 모든 값<for roop이 될때 마다 z 값은 달라진다>
    # 이 z값을 inner product를 해준다. 
    
    # ex) 이를 기반으로 e^2π*(0,1,2, ....)/100*i 
    # ((0,...,2π)i -> 100개의 sample로 한바퀴를 돈것을 만들었음 
    # (0,......4π)i -> 100개의 sample로 두바퀴 돈것을 만들었음 -> 50개가지고 한 바퀴를 돌음 -> frequency가 더 높음 
    # n의 값이 들어가면 (0,.....2nπ)i -> 100개의 sample로 n바퀴 돈 것을 만듬 -> 여기서 n의 값은 100임 -> 1개 가지고 1바퀴 
    # 이 모든 것을 inner product를 해준다. 
    # 여기서는 n = 1일때, e^2π*(0,1,2,....,4999)/5000*i -> (0,....,2π) -> 5000개의 sample로 한 바퀴 돈 것 
    # n = 2 -> e^4π*(0,1,2,....,4999)/5000*i -> 5000개의 sample로 두 바퀴 돈 것 
    # 즉 주기가 n이고, 주파수가 T인 벡터 값이 z이다. 
    
    amp.append(np.abs(np.dot(s,z)))
    # 'abs'는 절대값을 구해주는 함수 
    # -> 절댓값은 그 길이를 알려줌 -> 실수이면 x축 상의 값이 니까 원점에서 그 점까지의 길이가 되느느 것임 
    # dot은 내적을 해주는 함수 
    # amp에 허수가 들어 있다고 하면 틀린 말이다. -> 절댓값을 해주었기 때문에 
    # 그래서 amp를 ploting하면 끝이다. 
    # append는 amp라는 벡터(행렬) 마지막에 append 뒤의 값을 집어 넣어 주는 것 
    # 이를 구하면 amp는 총 5000개가 나온다. -> 그리고 밑의 그래프에서 peak의 값이 된다. -> amp값이 높을 수록 그 성분이 많다는 뜻.
```

```python
def frame_signal(signal, srate, win_size, win_step):
    frame_size = int(win_size * srate)
    frame_step = int(win_step * srate)

    # Get number of frames
    num_frames = int(np.ceil(np.abs(len(signal) - frame_size) / frame_step))
    # Pad signal
    pad_len = num_frames * frame_step + frame_size
    pad = np.zeros((pad_len - len(signal)))
    signal_padded = np.append(signal, pad)
    # Get within-frame sample indices
    idx1 = np.tile(np.arange(0, frame_size), (num_frames, 1))
    # Get vectors of frame_step increments
    idx2 = np.tile(np.arange(0, num_frames * frame_step, frame_step),
                   (frame_size, 1)).T
    # Get total indices divided by each frame
    indices = idx1 + idx2
    # Get frames divided by each frame based on indices
    frames = signal_padded[indices.astype(np.int32, copy=False)]
#     if cutoff is not None:
#         cutoff_bin = round(cutoff * nfft / (srate))  # hz -> bin
#         frames = frames[:, :cutoff_bin]
    return frames

def get_window(win_size, srate, win_type, win_samp=None):
    '''Get window samples for win_size'''
    if win_samp is None:
        win_samp = int(win_size * srate)  # sec -> sample

    if win_type == 'rect':
        return np.kaiser(win_samp, 0)
    elif win_type == 'hamming':
        return np.hamming(win_samp)
    elif win_type == 'hanning':
        return np.hanning(win_samp)
    elif win_type == 'kaiser':
        return np.kaiser(win_samp, 14)
    elif win_type == 'blackman':
        return np.blackman(win_samp)

def plot_spectrogram(S):
    fig, ax = plt.subplots(facecolor='white', figsize=(14, 6))
    im = ax.imshow(S.T, aspect='auto', origin='lower', cmap=plt.get_cmap('Greys'))
    divider = make_axes_locatable(ax)
    cax = divider.append_axes('right', size='5%', pad=0.05)
    cbar = fig.colorbar(im, cax=cax)
    return fig, ax

def preemphasis(s, pre_emp=0.97):
    # Emphasize high frequency range of the waveform by increasing power(squared amplitude).
    s = lfilter([1, -pre_emp], [1], s)
    return s
```

```python
max_freq = None  # cutoff freq
win_size = 0.008 # sec
win_step = 0.001  # sec
win_type = 'hanning' # options: 'rect', 'hamming', 'hanning', 'kaiser', 'blackman'
nfft = 1024

# Emphasize signal
s = preemphasis(s)
# Frame signal
frames = frame_signal(s, sr, win_size, win_step)
# Apply window function
frames *= get_window(win_size, sr, win_type)
print('frames:', frames.shape)
# wave -> 한장짜리만 나올텐데 이를 어떻게 연결하여 만들까? 
# -> win size, win step -> 한 장 한 장 이동하면서 sepctrum을 만드는 방법 
# 한장의 길이 : win_size: 0.008초 -> 소리를 0.008초로 쪼개서 
# win step -> 0.001초 -> 마다 분석한뒤 
# 이를 이어 붙여 준다. 
```
